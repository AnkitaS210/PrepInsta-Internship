# Visualization

```python
plt.figure(figsize = (8, 6))
plt.ticklabel_format(style = 'plain')   # display numeric values in plain decimal format and not in log format
sns.histplot(cars_df['price'], kde = True)
plt.xlabel('price')
plt.title('Histogram Plot')
plt.show()
```

```python
# histogram

plt.figure(figsize = (7, 4))   # 7 = l, 4 = b
sns.histplot(cars_df['fuel_tank_capacity'], kde = True)
plt.title('Distribution of Fuel Tank Capacity')
plt.xlabel('Fuel Tank Capacity')
plt.ylabel('Frequency')
plt.show()
```

```python
# boxplot

plt.figure(figsize = (7, 4))
plt.ticklabel_format(style = 'plain')   # display numeric values in plain decimal format and not in log format
sns.boxplot(x = cars_df['price'])
plt.title('Boxplot of Prices')
plt.xlabel('Price')
plt.show()

# due to too much of data points we cannot see the box plot clearly
```

```python
plt.figure(figsize = (27, 7))
sns.countplot(x = 'company', data = cars_df, palette = 'viridis', order=cars_df['company'].value_counts().index)
plt.title('Count of Cars by Company')
plt.xlabel('Car Company')
plt.ylabel('Count')
plt.xticks(rotation = 55)
plt.show()
```

# Bivariate Analysis

```python
# heatmap
correlation_matrix = cars_df.corr()

# plotting the correlation map
plt.figure(figsize = (10, 7))
sns.heatmap(correlation_matrix, annot = True, cmap = 'BuPu', fmt = '0.2f', linewidths = 0.9)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize = (18, 15))
sns.scatterplot(x = 'price', y = 'cylinders', data = cars_df, hue = 'company', palette = 'viridis', s = 200)
plt.title('Price vs Cylinders')
plt.xlabel('Price')
plt.ylabel('Cylinders')
# plt.legend(title = 'Company')
plt.show()
```

```python
# pairplot

numeric_columns = ['fuel_tank_capacity', 'number_of_airbags', 'seating_capacity']

# creating a pair plot
sns.pairplot(cars_df[numeric_columns + ['model']], hue = 'model',  height = 5)
plt.suptitle('Pair plot of numeric variables', y = 1.5)

plt.show()
```

```python
# Get the top five body types based on total count
top_five_body_types = cars_df['body_type'].value_counts().head(5).index

# Filter the DataFrame to include only the top five body types
filtered_cars_df = cars_df[cars_df['body_type'].isin(top_five_body_types)]

# Order the sub-categories based on sales in descending order
order = cars_df.groupby('company')['price'].mean().sort_values(ascending=False).index

plt.figure(figsize=(28, 8))
sns.barplot(x='company', y='price', hue='body_type', order = order, data=filtered_cars_df, palette='Set1')
plt.title('Price by Company with Top Five Body Types')
plt.xlabel('Company')
plt.ylabel('Price')
plt.legend(title='Body Type')
plt.show()
```
