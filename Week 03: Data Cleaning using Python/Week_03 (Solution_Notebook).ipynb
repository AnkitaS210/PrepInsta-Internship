# Data Cleaning - 1

# column names
(converting from upper to lower case)
*   columns = to access all the columns
*   str = to convert the data to a string
*   lower = convert to lower case

```python
# changing col names to lowercase:
data.columns.str.lower()

# replacing ' ' with '_'
data.columns = data.columns.str.lower().str.replace(' ', '_')
```

```python
# updations

data.head()
```

# index column
(dropping this column as we already have index)

*   axis = 1 : indicating that the operation should be applied to columns
*   axis = 0 : indicating that the operation should be applied to rows

*   inplace = True : modifies the existing data frame and does not creates a new one

```python
# dropping the column named 'index'

data = data.drop('index', axis=1)
```

```python
# updations

data.head()
```

# age column:
(treating missing values in this column and making round values )

```python
# accessing age column (can be done using two methods)

# first method:
data['age']

# second method:
data.age
```

```python
# calculating mean

avg_age = data['age'].mean()
avg_age
```

```python
# filling missing values with calculated mean

data['age'] = data.age.fillna(avg_age)

# rounding up decimals

data['age'] = data.age.round().astype(int)
```

```python
# updations

data.head()
```

# salary column
(removing "$" ,
replacing "k" with "1000's" and
distributing salary into max and min)

```python
# removing/ replacing

data['salary'] = data.salary.str.replace('$', '', regex = True).str.replace('k', '000', regex = True)
```

```python
# splitting columns

data[['min_salary', 'max_salary']] = data.salary.str.split('-', expand=True)

# dropping the original 'salary' column
data = data.drop('salary', axis=1)
```

```python
# updations

data.head()
```

# rating column
(treating missing values, and replacing negative values with median)

```python
# finding median for treating missing values

r_median = data.rating.median()
r_median
```

```python
data['rating'] = data.rating.fillna(r_median).replace(-1.0, r_median)
```

```python
# updations

data.head()
```

# location column
(replacing In, Ny, Aus)

```python
data.location.value_counts()
```

```python
data['location'] = data.location.str.replace(',', '').str.rstrip('InNy Aus')
```

```python
# updations

data.head()
```

# established column
(replacing negative values with 'Unknown')

```python
data['established'] = data.established.replace(-1, 'Unknown')
```

```python
# updations

data.head()
```

# easy_apply column
(replacing False insted of -1)

```python
data['easy_apply'] = data.easy_apply.str.replace('-1', 'FALSE').str.title()
```

```python
# checking the updations

data.head()
```

# Data Cleaning - 2

```python
# checking missing values

df.isnull().sum()
```

# choice_description column
(treating missing values and splitting into different choices columns)

```python
# treating missing values

df['choice_description'] = df.choice_description.fillna('None')
```

```python
df['choice_description'] = df.choice_description.str.replace('[', '', regex = True).str.replace(']', '', regex = True)
```

```python
# updations
df.head()
```

```python
# splitting columns

choices = df.choice_description.str.split(', ', expand = True)

# Concatenate the new columns with the original DataFrame
df = pd.concat([df, choices.add_prefix('Choice_')], axis=1)
```

```python
df.head()
```

# item_price column
(replacing '$' with ' ')

```python
df['item_price'] = df.item_price.str.replace('$', '', regex = True)
```
