# Solving SQL Murder Mystery in Python

```python
# we need sqlite3 and pandas in order to run the sql queries in pandas

import pandas as pd
import sqlite3 as sql
```

```python
# setting up connection to the database

con = sql.connect ('/content/sql-murder-mystery.db')
```

```python
print(con)

# con represents connection
# since we are getting o/p without any error means the connection is working fine
```

remember: to add ; (semicolon) at the end of each query.

```python
# running our first query, looking for data from crime scene report

query_1 = '''
SELECT *
FROM crime_scene_report
WHERE city = 'SQL City'
ORDER BY date;
'''

# setting dataframe width to max
pd.set_option('display.max_colwidth', None)

# executing our query
pd.read_sql_query(query_1, con)
```

```python
# checking personal details of both the witnesses
# location : Northwestern Dr

query_2 = '''
SELECT *
FROM person
WHERE address_street_name = 'Northwestern Dr'
ORDER BY address_number desc;
'''
# desc is used for descending order if used nothing insted of desc by default it takes it as asc i.e. ascending

pd.read_sql_query(query_2, con) [:1]
# as we have seen in the data that the witness lives at the last house on "Northwestern Dr" so we used [0:1] in desc order
```

```python
# location : Franklin Ave

query_3 = '''
SELECT *
FROM person
WHERE name like '%Annabel%' AND address_street_name = 'Franklin Ave';
'''
pd.read_sql_query(query_3, con) [:5]
# [:5] is slicing used to only get the first 5 data
```

```python
# viewing the interview of both the witnesses taken after the murder

query_4 = '''
SELECT *
FROM interview
WHERE person_id = 14887 OR person_id = 16371;
'''
# above are the id's of both the witnesses

pd.read_sql_query(query_4, con) [:5]
```

Here, we got two clues:

1. The killer is a man and is a member of a gym with the status of gold and is having a membership number starting with '48Z' and left in a car with a number plate of 'H42W'

2. He was working out in the gym on 9th of Jan.



```python
# checking the gym database with above details

query_5  = '''
SELECT *
FROM get_fit_now_check_in
WHERE membership_id like '%48Z%' AND check_in_date = 20180109
ORDER BY check_in_date;
'''
pd.read_sql_query(query_5, con)
```

Two members were found with their membership id

```python
# now lets check the car details by the above details

query_6 = '''
SELECT *
FROM drivers_license
WHERE plate_number like '%H42W%';
'''
pd.read_sql_query(query_6, con)
```

Found two males with a plate number containing 'H42W'

```python
# checking personal details of both the males from above query

query_7 = '''
SELECT *
FROM person
WHERE license_id = '423327' OR license_id = '664760';
'''
pd.read_sql_query(query_7, con)
```

```python
# checking from above both who is the member of gym

query_8 = '''
SELECT *
FROM get_fit_now_member
WHERE name = 'Tushar Chandra' OR name = 'Jeremy Bowers' ;
'''
pd.read_sql_query(query_8, con)
```

We have finally found the murderer named **'Jeremy Bowers'**.
Both, the membership id as well as the membership status matches as per the information we found earlier.

So, the real villian is a woman with a Tesla car and red hair. Using the above clues lets find out who's the mastermind behind this murder.

```python
query_10 = '''
SELECT *
FROM drivers_license
WHERE car_make = 'Tesla' AND car_model = 'Model S' AND
gender = 'female' AND hair_color = 'red';
'''
pd.read_sql_query(query_10, con)
```

Here, we got three woman with Tesla Model S and red hair color

```python
# personal details of the above three woman

query_11 = '''
SELECT *
FROM person
WHERE license_id = '202298' OR license_id = '291182' OR license_id = '918773'
'''
pd.read_sql_query(query_11, con)
```

```python
# checking the event SQL symphony concert
query_12 = '''
SELECT person_id, count(*), event_name
FROM facebook_event_checkin
GROUP BY person_id
having count(*) = 3 AND event_name = 'SQL Symphony Concert' AND date like '%201712%';
'''
pd.read_sql_query(query_12, con)
```
